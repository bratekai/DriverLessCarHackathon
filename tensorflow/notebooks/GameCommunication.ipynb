## Game Communication
Basically start playing the game and run this script

```python
import numpy as np
import game_communication
from matplotlib.pyplot import imshow
import matplotlib.pyplot as plt
%matplotlib inline

comm = game_communication.GameTelemetry()
#comm = game_communication.GameTelemetry("10.45.68.156")
#comm = game_communication.GameTelemetry("10.45.65.58")
```

```python
comm.connect()
```

### Get Image from game

```python
imshow(comm.get_image(4096))
#plt.show()
```

### Get other telemetry data

```python
print(comm.get_game_data())
```

### Send some information to the game

```python
commands = [0.0, 1.0]
resp = comm.send_command(commands)
print(resp)
```

### Disconnect from server

```python
comm.disconnect()
```
